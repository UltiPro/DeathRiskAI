```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_parquet("./Y_trainval.parquet")
df.shape

# Counting the number of deaths
# 0 = survived, 1 = died
gender_counts = df["death"].value_counts()

# Plotting
gender_counts.plot(kind="bar", color=["green", "red"])
plt.title("Number of Deaths in the Train/Validation Set")
plt.xlabel("Status")
plt.ylabel("Number of patients")
plt.xticks(ticks=[0, 1], labels=["Survived", "Died"], rotation=0)
plt.tight_layout()
plt.show()
```
